```python
import os
import pandas as pd

# Define the directory path
directory_path = r'C:\Users\rahul\Downloads\mainfolder\mainfolder\product_descriptions'

# Initialize the final dictionary
file_data_dict = {}

# Iterate through all files in the directory
for filename in os.listdir(directory_path):
    # Get the full file path
    file_path = os.path.join(directory_path, filename)
    
    # Skip if it's not a file
    if not os.path.isfile(file_path):
        continue
    
    # Get the file name without extension
    file_name_without_ext = os.path.splitext(filename)[0]
    
    # Check file extension and process accordingly
    if filename.endswith('.csv'):
        # Load the CSV file
        data = pd.read_csv(file_path)
        # Convert the DataFrame to a dictionary
        data_dict = data.to_dict(orient='records')
    elif filename.endswith('.txt'):
        # Read the contents of the text file
        with open(file_path, 'r') as file:
            data_dict = file.read()
    else:
        # Skip unsupported file types
        continue
    
    # Add the file data to the dictionary
    file_data_dict[file_name_without_ext] = data_dict

# Print the final dictionary
print(file_data_dict)
```

```python
import os
import pandas as pd
import json

# Define the directory path
directory_path = r'C:\Users\rahul\Downloads\mainfolder\mainfolder\product_details'

# Initialize the final dictionary
file_data_dict = {}

# Iterate through all files in the directory
for filename in os.listdir(directory_path):
    # Get the full file path
    file_path = os.path.join(directory_path, filename)
    
    # Skip if it's not a file
    if not os.path.isfile(file_path):
        continue
    
    # Get the file name without extension
    file_name_without_ext = os.path.splitext(filename)[0]
    
    # Check file extension and process accordingly
    if filename.endswith('.csv'):
        # Load the CSV file
        data = pd.read_csv(file_path)
        # Convert the DataFrame to a dictionary
        data_dict = data.to_dict(orient='records')
    elif filename.endswith('.txt'):
        # Read the contents of the text file
        with open(file_path, 'r') as file:
            data_dict = file.read()
    elif filename.endswith('.json'):
        # Load the JSON file
        with open(file_path, 'r') as file:
            data_dict = json.load(file)
    else:
        # Skip unsupported file types
        continue
    
    # Add the file data to the dictionary
    file_data_dict[file_name_without_ext] = data_dict

# Print the final dictionary
print(file_data_dict)
```

```python
import pandas as pd

# Define the file path
file_path = r'C:\Users\rahul\Downloads\mainfolder\mainfolder\sales_data.csv'

# Load the CSV file
sales_data = pd.read_csv(file_path)

# User input for the SKU to delete
sku_to_delete = input("Enter the SKU to delete: ")

# Check if the SKU exists in the DataFrame
if sku_to_delete in sales_data['Product_SKU'].values:
    # Delete the record with the specified SKU
    updated_sales_data = sales_data[sales_data['Product_SKU'] != sku_to_delete]
    
    # Save the updated data back to the CSV file, overwriting the existing file
    updated_sales_data.to_csv(file_path, index=False)
    
    print(f"Record with SKU '{sku_to_delete}' has been deleted.")
else:
    print(f"SKU '{sku_to_delete}' not found in the file.")

# Display the updated data
print(updated_sales_data)
```

```python
import pandas as pd

# Define the file path
file_path = r'C:\Users\rahul\Downloads\mainfolder\mainfolder\sales_data.csv'

# Load the CSV file
sales_data = pd.read_csv(file_path)

# User input for SKU, day, and new sales value
sku_to_update = input("Enter the SKU to update: ")
day_to_update = input("Enter the day to update (e.g., Day1, Day2, etc.): ")
new_sales_value = int(input(f"Enter the new sales value for {day_to_update}: "))

# Check if the SKU and day exist in the DataFrame
if sku_to_update in sales_data['Product_SKU'].values and day_to_update in sales_data.columns:
    # Update the specific day sales value for the given SKU
    sales_data.loc[sales_data['Product_SKU'] == sku_to_update, day_to_update] = new_sales_value
    
    # Save the updated data back to the CSV file, overwriting the existing file
    sales_data.to_csv(file_path, index=False)
    
    print(f"Sales value for SKU '{sku_to_update}' on {day_to_update} has been updated to {new_sales_value}.")
else:
    print(f"Either SKU '{sku_to_update}' or day '{day_to_update}' not found in the file.")

# Display the updated data
print(sales_data)
```
